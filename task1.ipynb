# **Creating a histogram plot on categorical data**

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
file_path = '/content/drive/My Drive/Prodigy_Infotech/categorical_data.csv'
data = pd.read_csv(file_path)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
# finding mode of region
mode_region=data['Region'].mode()[0]
print(mode_region)
```

```python
# filling missing values in region column with mode_region value
data['Region'].fillna('Europe & Central Asia', inplace=True)
print(data)
```

```python
# finding mode of IncomeGroup
mode_IncomeGroup=data['IncomeGroup'].mode()[0]
print(mode_IncomeGroup)
```

```python
# filling missing values in IncomeGroup column with mode_IncomeGroup
data['IncomeGroup'].fillna('High income',inplace=True)
print(data)
```

```python
# filling missing values in SpecialNotes column with some text
data['SpecialNotes'].fillna('There is no data to display',inplace=True)
print(data)
```

```python
data.drop(columns=['Unnamed: 5'], inplace=True)
print(data)
```

```python
missing_values=data.isnull().sum()
print(missing_values)
```

```python
import  matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='IncomeGroup', y='Region')
plt.title('Region vs IncomeGroup')
plt.show()
```

# **Creating a Bar Plot on Continuous Data**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
file_path = '/content/drive/My Drive/Prodigy_Infotech/Continuous_data.csv'
data = pd.read_csv(file_path, skiprows=4, on_bad_lines='skip', sep=None, engine='python')
print(data)
```

```python
continuous_columns = data.select_dtypes(include=['float64', 'int64']).columns
print(f"Continuous Columns: {continuous_columns}")
```

```python
selected_column = continuous_columns[3]
```

```python
cleaned_data = data.dropna(subset=[selected_column])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_data[selected_column], kde=False, bins=30, color='blue')
plt.xlabel(f'{selected_column}')
plt.ylabel('Frequency')
plt.title(f'Bar Plot of {selected_column}')
plt.show()
```
